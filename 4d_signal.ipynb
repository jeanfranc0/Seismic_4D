```python
import numpy as np
import pandas as pd
import segyio
import os

files_dic = {
    "drms": "/datashell/datasets/3D_insights/dRMS_m1aMinusBs_PhaseShift-90.sgy",
    "IP": "/datashell/datasets/3D_insights/Taynah/Diff_IP_trend_Bs_M1a_FS_v4_TWT",
    "LR": "/datashell/datasets/3D_insights/LR_052024.sgy",
    "MR": "/datashell/datasets/3D_insights/MR_052024.sgy",
    "Model01_dAMP": "/datashell/datasets/3D_insights/Model001/delta_Amp_quad_model001",
    "Model01_dIP": "/datashell/datasets/3D_insights/Model001/delta_IP_model001",
    "Model77_dAMP": "/datashell/datasets/3D_insights/Model077/delta_Amp_quad_model077",
    "Model77_dIP": "/datashell/datasets/3D_insights/Model077/delta_IP_model077",
}

def segy_to_df(tipo):
    file_path = files_dic[tipo]
    save_path = f"./data/raw_dfs/{tipo}.csv"

    os.makedirs("./data/raw_dfs", exist_ok=True)

    with segyio.open(file_path, "r", ignore_geometry=True) as f:
        inlines = f.attributes(segyio.TraceField.INLINE_3D)[:]
        xlines = f.attributes(segyio.TraceField.CROSSLINE_3D)[:]
        times = f.samples
        traces = np.stack([f.trace[i] for i in range(f.tracecount)], axis=0)

    print(f"inlines shape: {inlines.shape}")
    print(f"xlines shape: {xlines.shape}")
    print(f"traces shape: {traces.shape}")

    n_traces, n_samples = traces.shape
    df = pd.DataFrame({
        "inline": np.repeat(inlines, n_samples),
        "xline": np.repeat(xlines, n_samples),
        "time": np.tile(times, n_traces),
        tipo: traces.flatten()
    })

    df.to_csv(save_path, index=False)
    print(f"DataFrame salvo em: {save_path}")
    
segy_to_df("Model01_dAMP")
segy_to_df("Model01_dIP")
segy_to_df("Model77_dAMP")
segy_to_df("Model77_dIP")
```
